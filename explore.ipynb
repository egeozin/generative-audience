```python
# Importing required libraries
import pandas as pd

# Loading the dataset
#file_path = '/mnt/data/upworthy-archive-confirmatory-packages-03.12.2020.csv'
upworthy_data = pd.read_csv(csv_path)

# Displaying the first few rows to understand the structure
upworthy_data.head()
```

```python
# Grouping the data by 'clickability_test_id' and aggregating the impressions and clicks
grouped_data = upworthy_data.groupby('clickability_test_id').agg(
    total_impressions=('impressions', 'sum'),
    total_clicks=('clicks', 'sum')
).reset_index()

# Merging the grouped data back to the original dataset
merged_data = upworthy_data.merge(grouped_data, on='clickability_test_id')

# Displaying the first few rows of the merged data
merged_data.head()
```

```python
# Sorting the rows based on total number of impressions
sorted_data = merged_data.sort_values(by='total_impressions', ascending=False)

# Generating a new column for normalized click scores
sorted_data['normalized_click_score'] = sorted_data['clicks'] / sorted_data['total_clicks']

# Selecting the relevant columns to match the requested format
final_data = sorted_data[['clickability_test_id', 'headline', 'impressions', 'clicks', 'total_impressions', 'normalized_click_score']]
final_data.columns = ['test_id', 'headline', 'impressions', 'clicks', 'total_test_id_impressions', 'click_score']

# Displaying the first few rows of the final data
final_data.head()
```

```python
# Finding the clickability_test_ids that have at least two unique headlines
test_ids_with_unique_headlines = final_data.groupby('test_id').filter(lambda x: x['headline'].nunique() >= 2)

# Saving the filtered data to a CSV file
filtered_csv_path = 'local_upworthy_filtered_data.csv'
test_ids_with_unique_headlines.to_csv(filtered_csv_path, index=False)

# Displaying the first few rows of the filtered data to confirm the changes
test_ids_with_unique_headlines.head(), filtered_csv_path
```
